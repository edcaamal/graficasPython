### Reconocimiento de Rostros de Gatos Ver 1-0

```python

import cv2
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt

#Cargamos nuestro classificador de gatos:
#cascada_rostro = cv2.CascadeClassifier('haarcascade_frontalcatface_extended.xml')

cascada_rostro = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalcatface.xml')
```

```python
#Cargamos la imagen y la convertimos a grises:
#img = cv2.imread('imagen_gatos.jpeg')

img = cv2.imread('/content/gatitos.jpg')

# img = cv2.imread('perro_gato.jpeg')
# img = cv2.imread('perro_gato2.jpeg')
# img = cv2.imread('perroygato.jpeg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
img_gris = cv2.imread('/content/gatitos.jpg')

plt.imshow(img)
plt.show()
```

```python

#Buscamos los gatitos:
coordenadas_rostros = cascada_rostro.detectMultiScale(img_gris, 1.1, 2)
print(coordenadas_rostros)
```

```python

#Marcamos todos los gatos de la imagen:
for (x,y,ancho, alto) in coordenadas_rostros:
    cv2.rectangle(img_gris, (x,y), (x+ancho, y+alto), (0,255,0) , 3)


```

```python

#Abrimos una ventana con el resultado:
cv2_imshow(img_gris)
cv2_imshow(img)
```
