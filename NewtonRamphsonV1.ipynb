## Método Newton-Rapshon
El Objetivo de este método es estimar la solución de una ecuación f(x) = 0 es para producir una sucesion de aproximación que se acerquen a la soluci{on (iteraciones). Escogemos el primer nivel X0 de la secuencia y luego en circunstancias favorabñles el método hace el resto moviendose paso a paso hacia la raíz.

Es necesario realizar la derivada de la funcion

procedimiento.PNG

```python
# Versión 1

from math import *

def f(x):
  return cos(x)-x**3

def df(x):
  return -sin(x)-3*pow(x,2)

def NewtonRamphson(x0, tol,n):
  # tol = tolerancia
  for k in range(n):
    x1 = x0- f(x0)/df(x0)
    if (abs(x1-x0)<tol):
      print('x', k+1, '=',x1,' es la mejor aproximación de la raíz')
      return
    x0 = x1 
    print('x', k+1, '=',x1)

NewtonRamphson(pi, 0.0000001, 10)
```
