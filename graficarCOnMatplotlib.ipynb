```python
unidad1 = np.random.randint(6, 10, lenAlumnos) 
unidad2 = np.random.randint(6, 10, lenAlumnos) 
print(unidad1)
print(unidad2)

# Ajustar eje Y con el Promedio del Arreglo

promedio = unidad1.sum()/lenAlumnos
print(promedio)

# Dot en el eje Y
plt.plot(alumnos, unidad1, label="Unidad 1", color = "orange", marker = "o", linestyle=":")
plt.plot(alumnos, unidad2, label="Unidad 2", color = "green", marker = "<", linestyle=":")

# Linea limite, con ajuste del promedio
y0 = np.zeros(lenAlumnos)+promedio
plt.plot(alumnos,y0, 'b', linewidth = 1.0)

# Activar cuadricula
plt.grid()

plt.show()
```

```python
limAprobado = 6.9
unidad1 = np.random.randint(6, 10, lenAlumnos) 
unidad2 = np.random.randint(6, 10, lenAlumnos) 

print(unidad1)
print(unidad2)

# CAlcular resultado Final
final = np.add(unidad1, unidad2)/2
print(final)

# Ajustar eje Y con el Promedio del Arreglo

promedio = unidad1.sum()/lenAlumnos
print(promedio)

# Dot en el eje Y
plt.plot(alumnos, unidad1, label="Unidad 1", color = "orange", marker = "o", linestyle=":")
plt.plot(alumnos, unidad2, label="Unidad 2", color = "green", marker = "<", linestyle=":")
plt.plot(alumnos, final, label="Final", color = "black", marker = "*", linestyle="-")

# Linea limite, con ajuste del promedio
y0 = np.zeros(lenAlumnos)+ limAprobado
y1 = np.zeros(lenAlumnos)+promedio
plt.plot(alumnos,y0, 'r', linewidth = 1.0)
plt.plot(alumnos,y1, 'k', linewidth = 0.5)

# Activar cuadricula
plt.grid()

plt.show()
```

```python
import numpy as np

annos = ["2011", "2012",  "2013", "2014", "2015", "2016", "2017", "2018", "2019", "2020"]

nivel1 = np.random.rand(10)*100
nivel2 = np.random.rand(10)*200 + 100
nivel3 = np.random.rand(10)*300 + 200
nivel4 = np.random.rand(10)*400 + 300
nivel5 = np.random.rand(10)*500 + 400

nivel1
```

```python
plt.plot(annos, nivel1)
plt.plot(annos, nivel2)
plt.plot(annos, nivel3)
plt.plot(annos, nivel4)
plt.plot(annos, nivel5)
plt.show()

```

```python
plt.plot(annos, nivel1, label="nivel 1")
plt.plot(annos, nivel2, label="nivel 2")
plt.plot(annos, nivel3, label="nivel 3")
plt.plot(annos, nivel4, label="nivel 4")
plt.plot(annos, nivel5, label="nivel 5")

# Activar leyndas
plt.legend()

# Titulo de la grafica
plt.title("Examen de certificación")

# Etiquetas de los ejes
plt.xlabel("Años Certificación")
plt.ylabel("Puntaje")

# Personalizar eje Y
plt.yticks(np.arange(0, 1001, 50))

# Activar cuadricula
plt.grid()

# Activar marcas menores
plt.minorticks_on()

plt.show()
```

```python
plt.plot(annos, nivel1, label="nivel 1", color = "purple", marker = "<", linestyle="-")
plt.plot(annos, nivel2, label="nivel 2", color = "green", marker = "*", linestyle="--")
plt.plot(annos, nivel3, label="nivel 3", color = "blue", marker = "o", linestyle=":")
plt.plot(annos, nivel4, label="nivel 4", color = "black", marker = ".", linestyle="-.")
plt.plot(annos, nivel5, label="nivel 5", color = "red", marker = "+", linestyle=" ")

plt.legend()
plt.title("Examen de certificación")
plt.xlabel("Años Certificación")
plt.ylabel("Puntaje")

plt.yticks(np.arange(0, 1001, 50))
plt.grid()
plt.minorticks_on()

plt.show()
```
