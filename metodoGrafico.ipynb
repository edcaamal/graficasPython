# Análisis numérico

En análisis numérico un algoritmo de búsqueda de raíces es un método numérico o algorítmico para encontrar las soluciones aproximadas de una ecuación dada por la expresión f(x) = 0 para una función matemática f dada. A la solución x de la ecuación se le llama raíz o cero de la función.​

​


​

El algoritmo más simple de búsqueda de raíces es el método de bisección.
El método de Newton
El método de la regla falsa 


Método gráfico

```python
# Programa para realizar una grafica de una funcion para estimar raices

def f(x):
  f = x**5 -12.1*x**4 + 40.59*x**3 - 17.05*x**2 - 71.95*x + 35.88
  return f

# def f(x):
# return -2*x + 5


 
```

```python
# Librerias
import numpy as np
import matplotlib.pyplot as plt

# Libreria para manipular el espacio de los ejes
from matplotlib.ticker import AutoMinorLocator, MultipleLocator

# Solicita al usuario los valores del intervalo
xini = float(input("Valor inicial : ")) 
xfin = float(input("Valor final : ")) 

# Test -2.0 y 4.0
# Test 0, 10
# Test para grafica -1.5 y 2.5

# Test 2.5, 3.5
print(xini, xfin)
```

```python
# Genera la tabulacion
npart = 100
x = np.linspace(xini, xfin, npart+1)
print(x)
```

```python
# Genera la tabulacion para y
y = f(x)
print(y)
```

```python
# Realizar grafica dela funcion
# Grafica 1
plt.figure()
ax = plt.gca()
plt.gca
plt.plot(x, y, '-r', linewidth = 1.0)
plt.grid(True)
plt.show()
```

```python
# Realizar grafica dela funcion
# Grafica 2
plt.figure()
ax = plt.gca()
plt.gca
plt.plot(x, y, '-r', linewidth = 2.0)

# Espaciado de las particiones x, y
ax.xaxis.set_major_locator(MultipleLocator(0.25))
ax.xaxis.set_minor_locator(AutoMinorLocator(4))


plt.grid(True)
plt.show()
```

```python
# Realizar grafica dela funcion
# Grafica 3
plt.figure()
ax = plt.gca()
plt.gca
plt.plot(x, y, '-r', linewidth = 2.0)

# Graficar x = 0
y0 = np.zeros(len(x))
plt.plot(x,y0, 'b', linewidth = 1.0)

# Espaciado de las particiones x, y
ax.xaxis.set_major_locator(MultipleLocator(0.25))
ax.xaxis.set_minor_locator(AutoMinorLocator(4))


plt.grid(True)
plt.show()
```
