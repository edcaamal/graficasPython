```python
# Objeto con posiciones x,y, y color
import time

class figura:
  def __init__(self, lados, ancho, x, y, miColor):
    self.lados = lados
    self.ancho = ancho
    self.x = x
    self.y = y
    self.miColor = miColor

  def __str__(self):
    return f"{self.lados}.{self.ancho}.{self.x}.{self.y}.{self.miColor}"


  def dibujarFigura(self):
    up()
    setposition(self.x, self.y)
    down()
    color(self.miColor)
    grados = 360 / self.lados
    for i in range(self.lados):
      forward(self.ancho)
      right(grados)
    time.sleep(0.4)

  def moverFigura(self, x, y, apagar):
    if not (apagar):
       tmpColor = self.miColor
       self.miColor = 'black'
       figura.dibujarFigura(self)
       self.miColor = tmpColor
    self.x = x
    self.y = y
    figura.dibujarFigura(self)
```

```python
initializeTurtle(initial_speed=13)
triangulo = figura(3, 50, 100, 100, 'yellow')
cuadrado = figura(4, 50, 200, 100, 'red')
pentagono = figura(5, 50, 300, 100, 'orange')
hexagono = figura(6, 50, 400, 100, 'green')
heptagono = figura(7, 50, 500, 100, 'magenta')
octagono = figura(8, 50, 650, 100, 'white')
dodecagono = figura(12, 50, 300, 300, 'white')

triangulo.dibujarFigura()
cuadrado.dibujarFigura()
pentagono.dibujarFigura()
hexagono.dibujarFigura()
heptagono.dibujarFigura()
octagono.dibujarFigura()
dodecagono.dibujarFigura()
```

```python
initializeTurtle(initial_speed=13)
cuadrado = figura(6, 30, 150, 200, 'yellow')
cuadrado.dibujarFigura()
for i in range(20):
   cuadrado.moverFigura(150+(i*30), 200, False)
```
